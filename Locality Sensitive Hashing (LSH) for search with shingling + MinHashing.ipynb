```python
a = "flying fish flew by the space station"
b = "we will not allow you to bring your pet armadillo along"
c = "he figured a few sticks of dynamite were easier than a fishing pole to catch fish"
```

```python
def shingle(text: str, k: int):
  shingle_set = []
  for i in range(len(text) - k + 1):
    shingle_set.append(text[i:i+k])

  return set(shingle_set)
```

```python
k = 2
a = shingle(a, k)
b = shingle(b, k)
c = shingle(c, k)
```

```python
print(a)
```

```python
# find union to all shingling set
vocab = a.union(b).union(c)
```

```python
vocab
```

```python
vocab = list(vocab)
print(vocab)
```

```python
a
```

## one-hot encoding

```python
# one-hot encoding
a_1hot = [1 if x in a else 0 for x in vocab]
b_1hot = [1 if x in b else 0 for x in vocab]
c_1hot = [1 if x in c else 0 for x in vocab]
print(a_1hot)
print(b_1hot)
print(c_1hot)
```

## minHashing
convert aparse vectors --> dense vectors

```python
hash_ex = list(range(1, len(vocab) + 1))
print(hash_ex)
```

```python
from random import shuffle

shuffle(hash_ex)
print(hash_ex)
```

```python
for i in range(1, 10):
  print(f"{i} -> {hash_ex.index(i)}")
```

```python
for i in range(1, len(vocab) + 1):
  idx = hash_ex.index(i)
  signature_val = a_1hot[idx]
  print(f"{i} -> {idx} -> {signature_val}")
  if signature_val == 1:
    print('match!')
    break
```

```python
# function for creating the hash vector/function
def create_hash_func(size: int):
  hash_ex = list(range(1, len(vocab) + 1))
  shuffle(hash_ex)
  return hash_ex

# function for building multiple minhash vectors
def build_minhash_func(vocab_size: int, nbits: int):
  hashes = []
  for _ in range(nbits):
    hashes.append(create_hash_func(vocab_size))
  return hashes

# we create 20 minhash vectors
minhash_func = build_minhash_func(len(vocab), 20)
```

```python
print(minhash_func)
print(len(minhash_func))
```

```python
# function fo crating our signatures (eg the metching)
def create_hash(vector: list):
  signature = []
  for func in minhash_func:
    for i in range(1, len(vocab) + 1):
      idx = func.index(i)
      signature_val = vector[idx]
      if signature_val == 1:
        signature.append(idx)
        break
  return signature

# now crate signatures
a_sig = create_hash(a_1hot)
b_sig = create_hash(b_1hot)
c_sig = create_hash(c_1hot)

# this are compress version(dence vectors) of sparse vectors
print(a_sig)
print(b_sig)
print(c_sig)
```

```python
def jaccard(x, y):
  return len(x.intersection(y)) / len(x.union(y))

jaccard(set(a_sig), set(b_sig)), jaccard(a, b)
```

## LSH

```python
def split_vector(signature, b):
  assert len(signature) % b == 0
  r = int(len(signature) / b)
  # split signature in b parts
  subvecs = []
  for i in range(0, len(signature), r):
    subvecs.append(signature[i: i+r])
  return subvecs
```

```python
band_a = split_vector(a_sig, 10)
band_a
```

```python
band_b = split_vector(b_sig, 10)
band_b
```

```python
band_c = split_vector(c_sig, 10)
band_c
```

```python
for b_rows, c_rows in zip(band_b, band_c):
  if b_rows == c_rows:
    print(f"Candidate pair: {b_rows} == {c_rows}")
    # we only need one band to match
    break
```
